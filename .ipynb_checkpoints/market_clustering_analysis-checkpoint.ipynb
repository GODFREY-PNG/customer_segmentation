# Customer Revenue Intelligence: RFM-Based Segmentation
## Identifying High-Value Segments to Maximize CLV and Reduce Churn Attrition

---

### Business Case

Undifferentiated customer engagement is one of the primary drivers of marketing waste and churn attrition. 
By segmenting customers using behavioral transaction data, we can target retention spend precisely
directing acquisition budgets toward high-CLV profiles and intervention budgets toward at-risk transactors.

This analysis applies **K-Means clustering** on RFM (Recency, Frequency, Monetary) features derived from 
transactional records. The output is a set of commercially actionable customer tiers, each with a defined 
engagement strategy and estimated revenue impact.

**Data Assumption:** All transaction records are assumed to reflect a 12-month rolling window. 
Currency values are denominated in USD unless stated otherwise.

---

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pickle
import warnings
from datetime import datetime

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.utils.validation import check_is_fitted

# Suppressing deprecation and future warnings to keep pipeline output clean
warnings.simplefilter(action="ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

## 1. Data Loading & Cleaning

Raw marketing campaign data carries structural noise that corrupts downstream segmentation — 
invalid categorical entries, missing income values, and demographic outliers all distort cluster 
boundaries if left unaddressed. This section standardizes the dataset before any features are derived.

**Data Assumption:** The dataset reflects a single marketing campaign cycle. 
Income is denominated in USD. All date fields use dayfirst formatting (DD/MM/YYYY).

---

```python
df = pd.read_csv('data/MARKETING CAMPAIGN DATA.csv')
df.head()
```

```python
df.info()
```

```python
print("Education values:")
print(df['Education'].value_counts())
print()
print("Marital Status values:")
print(df['Marital_Status'].value_counts())
```

```python
# 'Alone', 'Absurd', 'YOLO' are non-standard entries with no distinct behavioral meaning
df['Marital_Status'] = df['Marital_Status'].replace(
    {'Alone': 'Single', 'Absurd': 'Single', 'YOLO': 'Single'}
)
```

```python
df.isnull().sum()
```

```python
missing_income_pct = df['Income'].isnull().sum() / len(df) * 100
print(f"Missing income: {missing_income_pct:.2f}%")
```

### Income Distribution & Imputation Strategy

At ~1% missingness, listwise deletion would introduce negligible bias — but income in marketing 
datasets is typically right-skewed (a small number of high earners pull the mean upward). 
We confirm the distribution visually before selecting an imputation method.

**Technical Reference — Imputation Choice:**
Median imputation is preferred over mean imputation when the distribution is right-skewed, 
as the median is robust to extreme values. Mean imputation in a skewed distribution 
systematically overestimates the central tendency for the majority of the population.



```python
df["Income"].hist(bins=10)
plt.title('Income Distribution')
plt.xlabel('Income (USD)')
plt.ylabel('Frequency');
```

```python
# Median chosen over mean — income distribution is right-skewed (confirmed above)
median_income = df["Income"].median()
df["Income"] = df["Income"].fillna(median_income)

print(f"Median applied: {median_income:,.0f}")
print(f"Remaining nulls: {df['Income'].isnull().sum()}")
```

### Outlier Detection — Income

Extreme income values distort the StandardScaler normalization step and compress the variance 
of the majority of customers into a narrow band. We use a boxplot to identify the outlier 
boundary before applying a hard cap.


```python
fig = px.box(
    data_frame=df,
    x="Income",
    title="Income Distribution — Outlier Inspection"
)
fig.update_layout(xaxis_title="Income (USD)")
fig.show()
```

```python
# Incomes above $200K are statistically extreme and inconsistent with the campaign's target market
df = df[df['Income'] < 200_000]
```

## 2. Feature Engineering

Raw transactional columns are too granular for stable clustering — K-Means distance 
calculations are sensitive to feature count, and redundant columns introduce noise that 
blurs segment boundaries. We engineer five composite features that compress behavioral 
signal into interpretable RFM-aligned dimensions:

Feature | Business Meaning 

| TotalSpend | Monetary value — wallet share captured across all product categories |
| `TotalPurchases` | Frequency proxy — channel-agnostic purchase volume |
| `TotalCampaignsAccepted` | Engagement score — receptiveness to marketing stimuli |
| `CustomerTenure` | Loyalty signal — days since first recorded transaction |
| `EducationLevel` | Socioeconomic proxy — ordinal encoding of qualification tier |
| `Age` | Demographic anchor — derived from birth year to avoid temporal drift |

**Data Assumption:** `Dt_Customer` is the account creation date, treated as the start of the 
customer relationship. Tenure is calculated relative to today's date dynamically.

---

```python
# combine all product spend into one total column
df['TotalSpend'] = df[['MntWines','MntFruits','MntMeatProducts',
                        'MntFishProducts','MntSweetProducts','MntGoldProds']].sum(axis=1)

# collapse all purchase channels into one total
df['TotalPurchases'] = df[['NumWebPurchases','NumCatalogPurchases',
                            'NumStorePurchases']].sum(axis=1)

# sum accepted campaigns — measures customer engagement
df['TotalCampaignsAccepted'] = df[['AcceptedCmp1','AcceptedCmp2',
                                    'AcceptedCmp3','AcceptedCmp4','AcceptedCmp5']].sum(axis=1)
# map education to numeric — Basic=1 up to PhD=4
education_order = {'Basic': 1, '2n Cycle': 2, 'Graduation': 3, 'Master': 4, 'PhD': 4}
df['EducationLevel'] = df['Education'].map(education_order)
# convert join date to days since joining-tenure
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True)
df['CustomerTenure'] = (pd.Timestamp.now() - df['Dt_Customer']).dt.days
```

```python
# quick check to confirm all new features look reasonable
print(df[['TotalSpend', 'TotalPurchases', 'TotalCampaignsAccepted', 'CustomerTenure','EducationLevel']].describe())
```

```python
# convert birth year to age
current_year = datetime.now().year  # pulling current year dynamically, not hardcoding

df['Age'] = current_year - df['Year_Birth']

# sanity check the new column
print(df[['Year_Birth', 'Age']].head())
print("Age stats:", df['Age'].describe())
```

```python
# drop unrealistic ages — anything above 90 is likely a data entry error
df = df[df['Age'] <= 90]

# sanity check for new column
print(df[['Year_Birth', 'Age']].head())
print("Age stats:", df['Age'].describe())
```

```python
# RFM features form the base of the segmentation
# Recency — days since last purchase (lower is better)
# Frequency — total purchases (higher is better)
# Monetary — total spend (higher is better)
rfm_features = df[['Recency', 'TotalPurchases', 'TotalSpend']]
print(rfm_features.describe())
```

```python
# check unique value counts — columns with too many or too few unique values
# are useless for clustering and will be dropped
df.nunique()
```

```python
# drop columns that add no value:
# ID — unique per row, Z_CostContact/Z_Revenue — zero variance
# originals replaced by engineered features (Age, CustomerTenure, aggregates)
print("Columns before drop:", df.shape[1])
```

```python
# drop identifier, zero-variance and original columns
# already replaced by Age, CustomerTenure, EducationLevel and the aggregated totals
df = df.drop(columns=[
    "ID", "Year_Birth", "Z_CostContact", "Z_Revenue", "Dt_Customer",
    "Education",  
    "MntWines", "MntFruits", "MntMeatProducts", "MntFishProducts",
    "MntSweetProducts", "MntGoldProds",
    "NumWebPurchases", "NumCatalogPurchases", "NumStorePurchases",
    "AcceptedCmp1", "AcceptedCmp2", "AcceptedCmp3", "AcceptedCmp4", "AcceptedCmp5"
])
df.info()
```

```python
print("Columns after drop:", df.shape[1])
print("Remaining columns:", df.columns.tolist())
```

```python
# check variance across all numeric columns to guide feature selection`
numeric_cols = df.select_dtypes(include=['number']).columns
print("Total numeric columns:", len(numeric_cols))
print()
print("All variances sorted:")
print(df[numeric_cols].var().sort_values())
```

```python
# clear variance gap after EducationLevel — top 8 features carry the signal, bottom 6 are too flat
top_features = df[numeric_cols].var().sort_values().tail(8)
print(top_features)
```

```python
# extract top 8 feature names — only these go into the model
top_features = df[numeric_cols].var().sort_values().tail(8)
high_var_feature = top_features.index.tolist()
print("Selected features:", high_var_feature)
```

```python
# horizontal bar chart to confirm the 8 selected features have the highest variance
fig = px.bar(
    x=top_features,
    y=top_features.index,
    orientation="h",
    title="Customer Segmentation: High Variance Features")
fig.update_layout(xaxis_title="Variance", yaxis_title="Features")
fig.update_xaxes(type='log')
fig.show()
```

```python
# correlation check on the 8 model features
correlation = df[high_var_feature].corr()
correlation
```

```python
# heatmap to spot strongly correlated feature pairs`
plt.figure(figsize=(12, 8)) 
sns.heatmap(correlation, annot=True, fmt=".1f", cmap="coolwarm")
plt.title('Customer Features Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# TotalSpend, TotalPurchases and Income correlate strongly but each covers a different RFM dimension
# keeping all three — dropping any would lose business meaning
# standardization will reduce the impact of the overlap before modeling
```

```python
# build feature matrix from the 8 selected columns
X = df[high_var_feature]
print("X shape:", X.shape)
```

```python
# check feature scales before standardization
X_summary = X.aggregate(["mean", "std"])
X_summary
```

```python
# K-Means uses distance — features on larger scales dominate unfairly
# StandardScaler brings everything to mean=0, std=1
ss = StandardScaler()
X_scaled_data = ss.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled_data, columns=X.columns)
print("X_scaled shape:", X_scaled.shape)
X_scaled.head()
```

```python
# Check which features still have missing values before modeling
print("Missing values in X:")
print(X.isnull().sum()[X.isnull().sum() > 0])
print("Total NaNs:", X.isnull().sum().sum())
```

```python
# Confirming scaling worked — mean should be ~0, std should be ~1 for all features
X_scaled_summary = X_scaled.aggregate(["mean", "std"]).astype(int)
X_scaled_summary
```

```python
# Testing K from 2 to 12 to find the optimal number of customer segments
# Inertia (elbow method) and silhouette score together give a stronger decision
n_clusters = range(2, 13)
inertia_errors = []
silhouette_scores = []

for k in n_clusters:
    model = make_pipeline(StandardScaler(), KMeans(n_clusters=k, random_state=42))
    model.fit(X)
    inertia_errors.append(model.named_steps["kmeans"].inertia_)
    silhouette_scores.append(silhouette_score(X, model.named_steps["kmeans"].labels_))

print("Inertia:", inertia_errors)
print("Silhouette Scores:", silhouette_scores)
```

```python
# elbow plot — find where inertia stops dropping sharply to pick optimal K
fig = px.line(
    x=n_clusters,
    y=inertia_errors,
    title="K-Means Model: Inertia vs Number of Clusters"
)
fig.update_layout(xaxis_title="Number of Clusters (K)", yaxis_title="Inertia")
fig.show()
```

```python
# silhouette score — confirms cluster separation, used alongside the elbow plot
fig = px.line(
    x=n_clusters,
    y=silhouette_scores,
    title="K-Means Model: Silhouette Score vs Number of Clusters"
)
fig.update_layout(xaxis_title="Number of Clusters (K)", yaxis_title="Silhouette Score")
fig.show()
```

```python
# K=2 peaks on silhouette but too broad for business use
# K=3 balances a strong score (~0.35) with meaningful segment differentiation
```

```python
# Building final model with K=3 based on elbow + silhouette analysis
final_model = make_pipeline(
    StandardScaler(),
    KMeans(n_clusters=3, random_state=42)
)
final_model.fit(X)
print("Model trained successfully")
```

```python
# Sanity check — confirms the model was actually trained before we use it
check_is_fitted(final_model)
```

```python
# Cluster labels assign each customer to a segment (0, 1, or 2)
labels = final_model.named_steps["kmeans"].labels_
print("Number of customers per segment:")
print(pd.Series(labels).value_counts().sort_index())
```

```python
# mean feature values per cluster — reveals what each segment looks like
xgb = X.groupby(labels).mean()
xgb
```

```python
# label segments based on centroid profiles:
# 0 — Budget Shoppers: low spend, high web visits, rarely convert
# 1 — High Value Champions: highest spend and purchase frequency
# 2 — Mid-Tier Regulars: highest income but underengaged, strong upsell potential
segment_names = {
    0: "Budget Shoppers",
    1: "High Value Champions",
    2: "Mid-Tier Regulars"
}
print("Segment profiles:")
print(xgb.rename(index=segment_names))
```

```python
# grouped bar chart — average feature value per segment, log scale handles the range difference
fig = px.bar(
    xgb,
    barmode="group",
    title="Average Customer Features by Segment"
)
fig.update_layout(xaxis_title="Segment", yaxis_title="Average Value (log scale)")
fig.update_yaxes(type="log")
fig.show()
```

```python
# 8 features can't be plotted directly — PCA compresses them to 2D for visualization
# scaling first so no single feature dominates the components
X_scaled_for_pca = StandardScaler().fit_transform(X)

pca = PCA(n_components=2, random_state=42)
X_t = pca.fit_transform(X_scaled_for_pca)
X_pca = pd.DataFrame(X_t, columns=["PC1", "PC2"])

print("PC1 explains:", pca.explained_variance_ratio_[0].round(2))
print("PC2 explains:", pca.explained_variance_ratio_[1].round(2))
print("Total variance explained:", pca.explained_variance_ratio_.sum().round(2))
```

```python
# 57% variance retained in 2D — enough to confirm cluster separation visually
# clusters that look distinct here are genuinely distinct in the full 8D space
```

```python
# scatter plot of clusters in 2D — color separation confirms the model found distinct groups
fig = px.scatter(
    data_frame=X_pca,
    x="PC1",
    y="PC2",
    color=labels.astype(str),
    title="Customer Segments Visualized in 2D (PCA)",
    labels={"color": "Segment"}
)
fig.update_layout(xaxis_title="Principal Component 1", yaxis_title="Principal Component 2")
fig.show()
```

```python
# champions and budget shoppers separate cleanly along PC1
# mid-tier sits higher on PC2 — likely age or tenure driven
# some middle overlap is normal — separation is clear enough for distinct strategies
```

```python
# save model to disk for use in the FastAPI backend`
with open("market_segmentation_model.pkl", "wb") as f:
    pickle.dump(final_model, f)
print("Model saved successfully as market_segmentation_model.pkl")
```
